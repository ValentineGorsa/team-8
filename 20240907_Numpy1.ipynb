```python
import numpy as np
import random
```

```python
coin = ["Heads", "Tails"]  #one dimentional array or list
```

```python
coin
```

```python
number_of_throws = 1000 #size
```

https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html

```python
tosses = np.random.choice(coin, number_of_throws, replace=True, p=None)
tosses
```

```python
print(f"The result of coin toss simulation for {number_of_throws} throws are {tosses}")
```

```python
#number of heads
heads_count = np.sum(tosses == "Heads")
heads_count
```

```python
#number of tails
tails_count = np.sum(tosses == "Tails")
tails_count
```

```python
#probalities
heads_prob = heads_count/number_of_throws
tails_prob = tails_count/number_of_throws
print(f'Probability of heads: {heads_prob}')
print(f'Probability of tails: {tails_prob}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100000 random coordinates
coordinates3 = np.random.randn(100000, 2)

x = coordinates3[:, 0] # X values
y = coordinates3[:, 1] # Y values

# Create a color range based on the y-values or any other function of coordinates
colors = np.arctan2(y, x)  # This will create a circular color gradient
# Create scatter plot with rainbow colors
plt.scatter(x, y, c=colors, cmap='rainbow', alpha=0.7)
plt.title('100000 randomly selected coordinates - normal distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()  # Adds a color bar for reference
plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Define the number of trials (tosses)
num_trials = 1000

# Initialize counters for each outcome
outcomes = {"HH": 0, "HT": 0, "TH": 0, "TT": 0}

# Simulate coin tosses
for _ in range(num_trials):
    coin1 = random.choice(["H", "T"])  # First coin
    coin2 = random.choice(["H", "T"])  # Second coin
    outcome = coin1 + coin2            # Combine results for two coins
    outcomes[outcome] += 1             # Record the outcome

# Calculate probabilities
probabilities = {outcome: count / num_trials for outcome, count in outcomes.items()}

# Display results
print("Outcome Frequencies:", outcomes)
print("Outcome Probabilities:", probabilities)

# Visualize with a bar chart
plt.bar(outcomes.keys(), outcomes.values(), color=['blue', 'green', 'orange', 'red'])
plt.title('Coin Toss Simulation (1000 trials)')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()

# Visualize probabilities with a pie chart
plt.pie(probabilities.values(), labels=probabilities.keys(), autopct='%1.1f%%', colors=['blue', 'green', 'orange', 'red'])
plt.title('Outcome Probabilities')
plt.show()
```
